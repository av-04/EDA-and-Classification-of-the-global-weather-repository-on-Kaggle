```python
# Create a copy to avoid modifying the original
df_fixed = df.copy()

# List of columns to fix
columns_to_fix = ['temperature_celsius', 'precip_mm']

print("--- Outlier Detection and Fixing Report ---")

# Store bounds for later use in plotting
bounds_info = {}

for col in columns_to_fix:
    print(f"\nProcessing column: {col}")
    
    # 1. Calculate IQR and Bounds
    Q1 = df_fixed[col].quantile(0.25)
    Q3 = df_fixed[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    
    # Store bounds
    bounds_info[col] = {'lower': lower_bound, 'upper': upper_bound}
    
    print(f"  Q1: {Q1:.4f}, Q3: {Q3:.4f}, IQR: {IQR:.4f}")
    print(f"  Calculated Lower Bound: {lower_bound:.4f}")
    print(f"  Calculated Upper Bound: {upper_bound:.4f}")
    
    # 2. Detect Outliers
    outliers_low = df_fixed[col] < lower_bound
    outliers_high = df_fixed[col] > upper_bound
    num_outliers = np.sum(outliers_low | outliers_high)
    
    print(f"  Outliers found: {num_outliers} (out of {len(df_fixed)})")
    
    # 3. Fix Outliers (Clipping)
    df_fixed[col] = np.clip(df_fixed[col], lower_bound, upper_bound)
    print(f"  Action: Values clipped to range [{lower_bound:.4f}, {upper_bound:.4f}]")

print("\n--- Verification: Summary Statistics After Fixing ---")
# Show .describe() for the fixed columns to prove it worked
print(df_fixed[['temperature_celsius', 'precip_mm']].describe())

# 4. Save the fixed data
df_fixed.to_csv('GlobalWeather_Fixed.csv', index=False)
print("\nFixed dataframe saved to 'GlobalWeather_Fixed.csv'")


# Create figure for "after" plots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Boxplot for Temperature (After)
sns.boxplot(y=df_fixed['temperature_celsius'], ax=ax[0])
ax[0].set_title('After: Temperature (Celsius) - Fixed')
ax[0].set_ylabel('Temperature (°C)')

# Boxplot for Precipitation (After)
sns.boxplot(y=df_fixed['precip_mm'], ax=ax[1])
ax[1].set_title('After: Precipitation (mm) - Fixed')
ax[1].set_ylabel('Precipitation (mm)')

plt.tight_layout()
plt.savefig('outliers_after_plots.png')

print("Generated 'after' outlier boxplots.")
```

```python

from sklearn.ensemble import IsolationForest


features = [
    'temperature_celsius', 'humidity', 'wind_kph',
    'pressure_mb', 'precip_mm', 'uv_index', 'cloud'
]
X = df[features].copy()

contamination_rate = 0.01
model = IsolationForest(
    n_estimators=100,
    contamination=contamination_rate,
    random_state=42,
    n_jobs=-1
)

df['anomaly'] = model.fit_predict(X)

df['anomaly_score'] = model.decision_function(X)

anomalies = df[df['anomaly'] == -1]
normal_data = df[df['anomaly'] == 1]

print(f"Total records: {len(df)}")
print(f"Number of anomalies detected (1% contamination): {len(anomalies)}")

top_anomalies = anomalies.sort_values(by='anomaly_score').head(5)

print("Top 5 Most Anomalous Weather Observations")

display_cols = ['country', 'location_name', 'last_updated'] + features + ['anomaly_score']
print(top_anomalies[display_cols].to_markdown(index=False, numalign="left", stralign="left"))

print("\n--- Comparison: Mean values of Anomalies vs. Normal Data ---")
comparison_df = pd.concat([
    normal_data[features].mean().rename('Normal Mean'),
    anomalies[features].mean().rename('Anomaly Mean')
], axis=1)

print(comparison_df.to_markdown(numalign="left", stralign="left"))
```
